```python
df = df.drop(['Survival Months'], axis=1)   # optional
```

```python
df.columns = df.columns.str.strip()
```

```python
# For Object / String Type
def fillnaObjectMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

object_cols = [
    'Race', 'Marital Status', 'T Stage', 'N Stage', '6th Stage',
    'differentiate', 'Grade', 'A Stage', 'Estrogen Status',
    'Progesterone Status', 'Status'
]

fillnaObjectMode(object_cols)

```

```python
# For Float / Integer Type
def fillnaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

numeric_cols = [
    'Age', 'Tumor Size', 'Regional Node Examined',
    'Reginol Node Positive'
]

fillnaFloatMean(numeric_cols)
```

```python
df.info()
```

```python



def convertObjToInt(cols):
    for col in cols:
        data = pd.DataFrame({col: df[col].unique()})
        le = LabelEncoder()
        le.fit(np.ravel(data))
        df[col] = le.transform(df[col])

object_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Object Columns:", object_cols)


convertObjToInt(object_cols)

```

```python

df = pd.read_csv('processed_data.csv')

df.columns
```

```python
X = df.drop(['Status'], axis=1)
y = df['Status']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False)
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model_rf = RandomForestClassifier()
model_rf.fit(train_X, train_y)
print(model_rf)

```

```python
model_pred_rf = model_rf.predict(test_X)
model_pred_rf
```

```python
from sklearn.metrics import accuracy_score
model_acc_rf = accuracy_score(test_y, model_pred_rf)
print(round(model_acc_rf,3) * 100, "%")
```
